```python
#another exercise
import json
import requests

url = "https://api.disneyapi.dev/character"
response = requests.get(url) #or requests.request("GET", url)
text = response.text
data = json.loads(text)

#char named Ambrose in which films?

for char in data["data"]:
    if char["name"] == "Ambrose":
        T = char["films"]

print("Ambrose appears in the following movies: \n")
for film in T:
    print(film)



#char --> most films and also how many?
counter = 0

for char in data["data"]:
    if len(char["films"]) > counter:
        counter = len(char["films"])
        charName = char["name"]

print("\n\nfollowing char has most films:", charName, "with", counter, "films")

print()

#provide a list of characters with at least 1 film and 1 videogame
L = []
for char in data["data"]:
    if char["videoGames"] != [] and char["films"] != []:
        L.append(char["name"])

print("characters with at least 1 film and 1 videogame")
for c in L:
    print(c)
```

## Webscraping

```python
import requests
import re

url = "https://en.wikipedia.org/wiki/Elephant"
response = requests.get(url)
if response.status_code == 200:
    html = response.text
    print(html)
    regex = "elephant|Elephant"
    matches = re.findall(regex,html)
    print(len(matches)) #--> count of elephant
else:
    print("try again")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html)
#print(soup)
#print(soup.prettify())
print(soup.title)
print(soup.title.get_text())
print(soup.get_text()[0:500].split())
print(soup.get_text().count("Elephant"))
print(soup.get_text().count("elephant"))

h3_tags = soup.find("h3") #returns first h3 tag found
print(h3_tags)

h3_tags = soup.find_all("h3")
print(h3_tags)

for element in h3_tags:
    print(element.get_text())
```

```python
#grabbing all of the languages on a wiki page
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
response = requests.get(url)
soup = BeautifulSoup(response.text)
li_tags = soup.find_all("li",{'class':"interlanguage-link"})
for tag in li_tags:
    print(tag.a["title"])
```

## more images and regex

```python
from bs4 import BeautifulSoup
import requests
import re

url = "https://en.wikipedia.org/wiki/Soup"
response = requests.get(url)
soup = BeautifulSoup(response.text)
img_tags = soup.find_all('img') #all img tags
src = [img['src'] for img in img_tags] #only the URL's

#using regex to get urls to stored images
result = re.findall(r'(/\w+.(jpg|gif|png))', str(src))

for url in result:
    print(url)
```
